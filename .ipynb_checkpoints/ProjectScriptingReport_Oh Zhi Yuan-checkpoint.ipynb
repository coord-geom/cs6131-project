### Overview

When planning for a holiday, one would encounter a plethora of choices on where to go, what hotels to stay in etc. Typically, those who want to think less about the nitty gritty details like the itinerary would turn to tour packages where almost everything is taken care of for you. Nevertheless, it is still a overwhelming task to decide on that best deal with a bag for your buck since there are a ton of options to choose from. VoyageVault aims to solve this issue. With just a few keywords and filters, you would be able to look through all the packages of your liking and make decisions faster with less hassle.

In the context of the project, VoyageVault consolidates information from various travel agencies and provide users with an all-in-one database. As there is much information being added, modified and removed from the system constantly, a database should be used. There is a need to retrieve information efficiently and for data to be reliable and properly maintained, so a RDBMS is preferred to other storage methods like text files.

<p>Stuff in <strong>bold</strong> are entities, <i>italics</i> are attributes.

Users of the website are called **customers** who will input their *first name*, *last name*, *email*, *phone number* and account *password*. Upon account creation, they will also be assigned a unique *customer id*.

Customers will query the database for **travel packages** which will have a unique *package id*, <s>*start date*, *end date*</s> <span style="background-color:#ffff00">*duration*</span>, *name*, *description*, *adult price*, *child price*, <span style="background-color:#ffff00">derived *average rating based on customer reviews*, *a link directed to the actual site for registration* and *zero or more tags* to denote the package's categories (e.g. exotic, mountain)</span>. Packages are offered by **travel agencies** which have a unique *agency id*, *name*, *hotline* and several *locations* for their offices. Agencies may offer none, one or many packages at any instant. There are two types of packages offered - **free-and-easy packages** and **tour packages**. Tour packages will have an *itinerary*, <s>while free-and-easy packages have a *recommendation* description</s>. <span style="background-color:#ffff00">These two are distinct and disjoint.</span> 

For tour packages, a customer would be part of a **tour group**, which has a <s>unique</s> *group id <span style="background-color:#ffff00">unique within the tour package</span>*, <span style="background-color:#ffff00">departure date</span> <s>*minimum group size*</s> <span style="background-color:#ffff00">*maximum group size* (determined by agency)</span> and *group size*. <s>We assume that there is no ceiling to the number of customers who can join a group.</s>

Customers on travel packages would go to one or more **countries** that will have a unique *country id* and *name* in the database. <s>They may depart from and arrive at different **airports** on certain *dates*. Airports have a unique *airport code* and name.</s> Their accomodation will be in **hotels** that have a <s>unique</s> *hotel id <span style="background-color:#ffff00">unique within the country</span>*, *name*, *address* and *rating*. <s>A country should have at least one airport and one hotel.</s>

<span style="background-color:#ffff00">Tour packages will visit **locations** in their trip, which have a *unique id within the country*, *name* and *region*, while free-and-easy packages will contain recommendations of locations.</span>
    
<span style="background-color:#ffff00">A customer can leave up to one review on a package. Each review would contain a *rating* and *the content*. A package can be reviewed by zero or more customers.</span>

<s>A customer will sign up for a package by making a **registration**. They would have to indicate the *number of adults* and *number of children* they sign up on behalf of. The *date* of registration is recorded and a unique *registration id* is assigned to it. Customers can register for multiple packages over time but should not be allowed to sign up for packages that overlap in dates.</s></p>

```python
%%sql
DROP DATABASE IF EXISTS VOYAGEVAULT;
CREATE DATABASE VOYAGEVAULT;
USE VOYAGEVAULT;

CREATE TABLE customer(
    cID integer,
    fname varchar(25),
    lname varchar(25),
    email varchar(64),
    hpnum integer,
    password varchar(64),
    primary key (cID)
);

CREATE TABLE agency(
    aID integer,
    aname varchar(64),
    hotline integer,
    primary key (aID)
);

CREATE TABLE agencyLoc(
    aID integer,
    location varchar(128),
    primary key (aID, location)
);

CREATE TABLE package(
    pID integer,
    aID integer,
    pname varchar(64),
    description varchar(512),
    duration integer,
    aprice integer,
    cprice integer,
    link varchar(128),
    itinerary varchar(2048),
    primary key (pID)
);

CREATE TABLE pTag(
    pID integer,
    tag varchar(64),
    primary key (pID, tag)
);

CREATE TABLE review(
    cID integer,
    pID integer,
    rating float,
    content varchar(1024),
    primary key (cID, pID)
);

CREATE TABLE country(
    countryID integer,
    cname varchar(64),
    primary key (countryID)
);

CREATE TABLE hotel(
    hID integer,
    countryID integer,
    hname varchar(64),
    haddr varchar(128),
    hrating float,
    primary key (hID, countryID)
);

CREATE TABLE location(
    lID integer,
    lname varchar(64),
    region varchar(64),
    countryID integer,
    primary key (lID, countryID)
);

CREATE TABLE tourGroup(
    gID integer,
    pID integer,
    maxsize integer,
    cursize integer,
    departure date,
    primary key (gID, pID)
);

CREATE TABLE accommodation(
    pID integer,
    hID integer,
    primary key (pID, hID)
);

CREATE TABLE packageDest(
    pID integer,
    countryID integer,
    primary key (pID, countryID)
);

CREATE TABLE visit(
    pID integer,
    lID integer,
    primary key (pID, lID)
);

CREATE TABLE recommend(
    pID integer,
    lID integer,
    primary key (pID, lID)
);
    
show tables;
```

#### Query 1

What tour groups are available so that I can bring a family of 11 for a trip?

```python
%%sql
SELECT p.pID, p.pname, p.duration, g.departure, g.maxsize - g.cursize available_capacity
FROM package p
JOIN tourGroup g ON p.pID = g.pID
WHERE g.maxsize - g.cursize >= 11
```

#### Query 2

Which packages has the most number of tags?

```python
%%sql
SELECT p.pID, pname, count(*) 'Number of tags'
FROM package p
INNER JOIN pTag ON p.pID = pTag.pID
GROUP BY p.pID
HAVING count(*) = (
  SELECT max(tagCount)
  FROM (
    SELECT count(*) as tagCount
    FROM pTag
    GROUP BY pID
  ) tagCounts
);
```

#### Query 3

What packages are there that go to Indonesia and costs $1500 or less for an adult?

```python
%%sql
SELECT p.pID, p.pname, p.aprice, p.cprice
FROM package p
INNER JOIN packageDest pd ON p.pID = pd.pID
INNER JOIN country c ON pd.countryID = c.countryID
WHERE c.cname = 'Indonesia' AND p.cprice <= 1500
ORDER BY p.cprice;
```

#### Trigger/Event

This trigger updates the average rating of a package after a new review on the package is made.

```python
%%sql
DROP TRIGGER IF EXISTS updatePRating;
CREATE TRIGGER updatePRating
AFTER INSERT ON review
FOR EACH ROW
BEGIN
  DECLARE avg_rating FLOAT;
  SELECT AVG(rating) INTO avg_rating FROM review WHERE pID = NEW.pID;
  UPDATE package SET prating = avg_rating WHERE pID = NEW.pID;
END;

show triggers;
```

This event runs daily to remove tour groups whose departure date has past.

```python
%%sql
DROP EVENT IF EXISTS removeTourGroupDaily;
CREATE EVENT removeTourGroupDaily
ON SCHEDULE EVERY 1 DAY
STARTS CURRENT_TIMESTAMP
DO
DELETE FROM tourGroup WHERE departure <= CURRENT_DATE();

show events;
```
